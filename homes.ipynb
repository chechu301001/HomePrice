```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import mpl_toolkits
%matplotlib inline
import seaborn as sns
```

```python
data = pd.read_csv("DS-Assignment-Part-1-dataset.csv")
data.head()
```

```python
data['Number of bedrooms']
```

```python
data['Number of bedrooms'].value_counts().plot(kind='bar')
plt.title('Number of Bedroom')
plt.xlabel('Bedrooms')
plt.ylabel('Count')
sns.despine()
```

So 1 Bedroom house is most common when compared to the rest

```python
plt.Figure(figsize=(10,10))
sns.countplot(x='Number of bedrooms', data=data)
```

```python
sns.jointplot(x=data.latitude.values, y=data.longitude.values, data=data, size=10)
plt.ylabel('Longitude', fontsize=12)
plt.xlabel('Latitude', fontsize=12)
plt.show()
```

These are the common Locations where houses are built

```python
plt.scatter(data['House price of unit area'], data['House size (sqft)'])
plt.title('House price of unit area vs House size (sqft)')
```

```python
plt.scatter(data['House price of unit area'], data['longitude'], color='red')
plt.title('House price of unit area vs longitude')
plt.xlabel('House price of unit area')
plt.ylabel('longitude')
```

```python
plt.scatter(data['House price of unit area'], data['latitude'], color='green')
plt.xlabel('House price of unit area')
plt.ylabel('latitude')
```

```python
plt.scatter(data['Number of bedrooms'], data['House price of unit area'], color='red')
plt.xlabel('Number of bedrooms')
plt.ylabel('House price of unit area')
sns.despine()
```
